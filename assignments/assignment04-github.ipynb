```python
import os
from dotenv import load_dotenv
from github import Github

# Load variables from .env file
load_dotenv()

# Access token securely
TOKEN = os.getenv("GITHUB_TOKEN")
REPO_NAME = os.getenv("REPO_NAME")

# Connect to GitHub
g = Github(TOKEN)
repo = g.get_repo(REPO_NAME)

# Get file
file = repo.get_contents("data/andrew.txt")
content = file.decoded_content.decode("utf-8")
```

```python
# Replace every occurrence of "Andrew" with "Aldas"
new_content = content.replace("Andrew", "Aldas")
```

```python
# Push update
repo.update_file(
    path="data/andrew.txt",
    message="Update file via Python",
    content=new_content,
    sha=file.sha,
    branch="main"
)

print("Done!")
```
