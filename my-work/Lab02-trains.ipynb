```python
import requests
from xml.dom.minidom import parseString  # fixed import

# URL for the real-time Irish Rail XML data
url = "http://api.irishrail.ie/realtime/realtime.asmx/getCurrentTrainsXML"

# Fetch the XML page
page = requests.get(url)

# Parse the XML content
doc = parseString(page.content)

# Check it works by printing a pretty XML to console
# print(doc.toprettyxml())  # comment this out once you know it works

# If you want to store the XML in a file
with open("trainxml.xml", "w", encoding="utf-8") as xmlfp:
    doc.writexml(xmlfp)
```

```python
# Get all <objTrainPositions> nodes
objTrainPositionsNodes = doc.getElementsByTagName("objTrainPositions")

# Loop through each node
for objTrainPositionsNode in objTrainPositionsNodes:
    # Get the first <TrainCode> child
    traincodenode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
    
    # Extract the text value and strip whitespace
    traincode = traincodenode.firstChild.nodeValue.strip()
    
    # Print the train code
    print(traincode)
```

```python
import csv

# Open the CSV file for writing (fixes blank lines issue)
with open('week03_train.csv', mode='w', newline='') as train_file:
    train_writer = csv.writer(
        train_file, 
        delimiter='\t', 
        quotechar='"', 
        quoting=csv.QUOTE_MINIMAL
    )
    
    # Get all <objTrainPositions> nodes
    objTrainPositionsNodes = doc.getElementsByTagName("objTrainPositions")
    
    # Loop through each node and write TrainCode to CSV
    for objTrainPositionsNode in objTrainPositionsNodes:
        traincodenode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
        traincode = traincodenode.firstChild.nodeValue.strip()
        train_writer.writerow([traincode])
        dataList = [] 
        dataList.append(traincode) 
        train_writer.writerow(dataList) 
```

```python
import csv

retrieveTags=[
    'TrainStatus', 'TrainLatitude', 'TrainLongitude', 
    'TrainCode', 'TrainDate', 'PublicMessage', 'Direction'
    ] 

# Open the CSV file for writing (fixes blank lines issue)
with open('week03_train.csv', mode='w', newline='') as train_file:
    train_writer = csv.writer(
        train_file, 
        delimiter='\t', 
        quotechar='"', 
        quoting=csv.QUOTE_MINIMAL
    )
    
    # Get all <objTrainPositions> nodes
    objTrainPositionsNodes = doc.getElementsByTagName("objTrainPositions")
    
    # Loop through each node and write TrainCode to CSV
    for objTrainPositionsNode in objTrainPositionsNodes:
        traincodenode = objTrainPositionsNode.getElementsByTagName("TrainCode").item(0)
        traincode = traincodenode.firstChild.nodeValue.strip()
        train_writer.writerow([traincode])
        dataList = [] 
        for retrieveTag in retrieveTags: 
            datanode = objTrainPositionsNode.getElementsByTagName(retrieveTag).item(0) 
            dataList.append(datanode.firstChild.nodeValue.strip())  
        train_writer.writerow(dataList) 
```
