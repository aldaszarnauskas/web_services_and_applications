```python
from google import genai
from dotenv import load_dotenv
import os
# The client gets the API key from the environment variable `GEMINI_API_KEY`.
load_dotenv()


client = genai.Client()

response = client.models.generate_content(
    model="gemini-2.5-flash", 
    contents="""Please generate 30 sentences in {A1} {Lithuanian} in a dialogue form.
    In a form like this:

    Title: [Place the title here]

    1. Text in lithuanian [Translation in english]
    2. Text in lithuanian [Translation in english]
    ...
    """
)
print(response.text)
```
