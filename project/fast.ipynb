```python
from google import genai
from dotenv import load_dotenv
import os
import re
# The client gets the API key from the environment variable `GEMINI_API_KEY`.
load_dotenv()

def get_a_dialogue(primary_lang, cambridge_level, second_lang, dialogue_length):

    client = genai.Client()

    primary_lang = "Lithuanian"
    cambridge_level = "A1"
    second_lang = "English"
    dialogue_length = 30

    response = client.models.generate_content(
        model="gemini-2.5-flash", 
        contents=f"""Please generate {dialogue_length} sentences in {cambridge_level} {primary_lang} in a dialogue form.
        In a form like this:

        1. Name: Text in {primary_lang} [Translation in {second_lang}]
        2. Name: Text in lithuanian [Translation in english]
        ...
        """
    )


    lines = response.text.split('\n')
        
    dialogue_lines = []
        
    for line in lines:
        line = line.strip()
        if not line:
            continue

        # Match: "1.  Jonas: Labas! [Hello!]"
        match = re.match(r'(\d+)\.\s+(\w+):\s+(.+?)\s*\[(.+?)\]', line)
        #print(line)
        if match:
            #print(match)
            dialogue_lines.append({
                "id": int(match[1]),
                "name": match[2].strip(),
                "primary_lang": match[3].strip(),
                "secondary_lang": match[4].strip()
            })
    
    return dialogue_lines
```
